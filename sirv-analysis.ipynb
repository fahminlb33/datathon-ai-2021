```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

plt.style.use("plot_style.txt")
```

```python
df = pd.read_csv("sim_summary.csv")
df
```

```python
#df_filtered = df[~df["simulation"].str.contains("s1")]
df_filtered = df.copy()
x = df_filtered["inf_ratio"].values * 100
y = df_filtered["simulation"].values
values = df_filtered["total_inf"].values

fig, ax = plt.subplots()
ax.barh(y, x)
for i, v in enumerate(values):
    ax.text(x[i] + 1, y[i], f"{values[i]} ({x[i]:.2f}%)")

ax.xaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_title("Infection Ratio")
```

```python
#df_filtered = df[~df["simulation"].str.contains("s1")]
df_filtered = df.copy()
df_filtered["inf_raise"] = (df_filtered["total_inf"] - df_filtered["inf_initial_count"]) / df_filtered["total_inf"]
x = df_filtered["inf_raise"].values * 100
y = df_filtered["simulation"].values
values = df_filtered["inf_raise"].values * 100

fig, ax = plt.subplots()
ax.barh(y, x)
for i, v in enumerate(values):
    ax.text(x[i] + 1, y[i], f"{values[i]:.2f}%")

ax.xaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_title("Infection Ratio")
```
